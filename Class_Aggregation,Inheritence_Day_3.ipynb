#Class relationships are of 2 types

1 : aggregation - called - Has-A

2 : Inheritence - called - Is-A

```python
#Aggregation of class

class Customer_profile():
  def __init__(self,name,age,gender,address):
    self.name=name
    self.age=age
    self.gender=gender
    self.address=address

class address():
  def __init__(self,city,state,country):
    self.city=city
    self.state=state
    self.country=country
```

```python
ad=address("patiala","punjab","india")

profile=Customer_profile("Rohan",23,"M",ad)


print(profile.address)

print(profile.address.city,profile.address.country)
```
