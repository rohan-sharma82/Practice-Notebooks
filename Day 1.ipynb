```python
class ATM():
  def __init__(self):
    self.pin=""
    self.balance=60000
    self.menu()

  def menu(self):
    while True: # Loop indefinitely until user exits
      user_input=input('''
            1: create your pin
            2: deposit money
            3: withdraw money
            4: check balance
            5: want to exit
      ''')

      if user_input=="1":
        print("Want to create ur pin")
        self.create_pin()

      elif user_input=="2":
        print("You have too much in cash")
        self.deposit()


      elif user_input=="3":
        print("you are out of cash ")
        self.withdraw()
      elif user_input=="4":
        print("You wanna know your worth.. hnnnnn")
        self.balance1()
      elif user_input=="5": # Exit condition
        print("Thank you for using the ATM, buddy!")
        break # Exit the loop
      else:
        print("Invalid option, please choose between 1-5.")


  def create_pin(self):
    self.pin=input("Enter Ur Pin- ")
    print("Pin set successfully")

  def deposit(self):
    temp= input("verify yourself with PIN -")
    if temp==self.pin:
      amount=int(input("enter the deposit amount- "))
      self.balance=self.balance + amount
      print("Yeah, I got it. Your worth is increased buddy")
    else:
      print("Invalid PIN")

  def withdraw(self):
    temp= input("enter your pin- ")
    if temp==self.pin:
      amount = int(input("Enter amount you want - "))
      if amount<=self.balance:
        self.balance=self.balance-amount
        print("withdraw successfull")
      else:
        print("buddy you don't even have that much money")
    else:
      print("Invailid")

  def balance1(self):
    temp= input("enter your pin- ")
    if temp==self.pin:
      print(self.balance)
    else:
      print("wrong pin")
```
