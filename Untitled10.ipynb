```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2)

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot([1, 2, 3], [1, 2, 3])
ax[0, 0].set_title("Top left")

ax[0, 1].plot([1, 2, 3], [3, 2, 1])
ax[0, 1].set_title("Top right")

ax[1, 0].scatter([1, 2, 3], [2, 4, 1])
ax[1, 0].set_title("Bottom left")

ax[1, 1].bar(["A", "B", "C"], [3, 7, 2])
ax[1, 1].set_title("Bottom right")

plt.tight_layout()
plt.show()
```
